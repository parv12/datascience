```python
from sklearn.pipeline import Pipeline
##feature Scaling
from sklearn.preprocessing import StandardScaler ## tran
from sklearn.linear_model import LogisticRegression ## estimator
```

```python
steps=[("standard_scaler",StandardScaler()),
       ("classifier",LogisticRegression())]
```

```python
steps
```

```python
pipe=Pipeline(steps)
```

```python
### visualize Pipeline
from sklearn import set_config
```

```python
set_config(display="diagram")
```

```python
pipe
```

```python
### creating a dataset
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size=0.33,random_state=42)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.decomposition import PCA 
from sklearn.svm import SVC
```

```python
step=[("scaling",StandardScaler()),
      ("PCA",PCA(n_components=3)),
      ("SVC",SVC())]
```

```python
pipe2['scaling'].fit_transform(X_train)
```

```python
pipe2.fit(X_train,y_train)
```

```python
pipe2.predict(X_test)
```

### complex examples of columns transformer

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
## num
numeric_processor=Pipeline(steps=[("imputation_mean",SimpleImputer(missing_values=np.nan,strategy="mean")),
                                  ("scaler",StandardScaler())])
```

```python
numeric_processor
```

```python
##categorical Processing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor=Pipeline(steps=[("imputation_constand",SimpleImputer(fill_value="missing",strategy="constand")),
                                  ("onehot",OneHotEncoder(handle_unknown="ignore"))])
```

```python
categorical_processor
```

```python
### 
```

```python
##combine processing technique
from sklearn.compose import ColumnTransformer
```

```python
preprocessor =ColumnTransformer([("categorical",categorical_processor,["gender","City"]),
                   ("numerical",numeric_processor,["age","height"])])
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe=make_pipeline(preprocessor,LogisticRegression())
```
